```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sns
```

```python
abprob = pd.read_csv("abprob.csv")
abprob.describe()
```

```python
great = abprob.loc[(abprob['prob'] >= .04999) & (abprob['prob'] <= 0.5001)]
good = abprob.loc[(abprob['prob'] >= .499) & (abprob['prob'] <= 0.501)]
okay = abprob.loc[(abprob['prob'] >= .03) & (abprob['prob'] <= .07)]
bad = abprob.loc[(abprob['prob'] >= .01) & (abprob['prob'] <= .09)]
#x = np.reshape(abprob['a'],(600,600))

plt.scatter(bad['a'],bad['b'],label = "Bad: .1-.9")
plt.scatter(okay['a'],okay['b'],label = "Okay: .3-.7")
plt.scatter(good['a'],good['b'],label = "Good: .499-.501")
plt.scatter(great['a'],great['b'],label = "Great: .4999-.5001")
plt.legend(loc = 'upper left')
plt.xlabel("Apha")
plt.ylabel("Beta")
plt.title("Beta Parameter Selection")
```
